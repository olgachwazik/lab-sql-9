### Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

```python
# joining two dataframes:
```

```python
rentals_5_and_6 = rentals_5.merge (rentals_6, on='customer_id')
rentals_5_and_6
```

```python
# creating the function to add column that will state in which month given customer borrowed more or less in june
```

```python
def may_vs_june(x):
    if x['count_may'] > x['count_june']:
        return 'less in june'
    elif x['count_may'] < x['count_june']:
        return 'more in june'
    elif x['count_may'] == x['count_june']:
        return 'same as may'
rentals_5_and_6['may_vs_june'] = rentals_5_and_6.apply(may_vs_june, axis=1)
```
